```python
import pandas as pd
```

```python
df = pd.read_csv('Sold_Updated.csv')
```

```python
df['SoldDate'] = pd.to_datetime(df['SoldDate'], errors='coerce', format='%m/%d/%Y')

# Convert date column to datetime format with YYYY-MM-DD format
df['SoldDate'] = pd.to_datetime(df['SoldDate'], errors='coerce')

df
```

```python
df = df.dropna()
df
```

```python
#getting rid of extra chandler zipcodes & keeping only one zip code per city
zipcodes_to_keep = [85225, 97229, 11368, 33186, 90210, 60629]
df = df[df['Zip'].isin(zipcodes_to_keep)]
df
```

```python
#arranging order of columns
df = df[['Address', 'City','Zip','SoldDate',
         'Bathrooms','Bedrooms','LivingArea','LandArea','PropPrice']] 
df
```

```python
df['Zip'] = df['Zip'].astype(str)
count_value_Portland = df['Zip'].value_counts()['97229']
count_value_NY = df['Zip'].value_counts()['11368']
count_value_Miami = df['Zip'].value_counts()['33186']
count_value_LA=df['Zip'].value_counts()['90210']
count_value_Chicago = df['Zip'].value_counts()['60629']
count_value_Chandler = df['Zip'].value_counts()['85225']
```

```python
#Getting a count of the property data per city
counts = {
    'Portland': count_value_Portland,
    'NY': count_value_NY,
    'Miami': count_value_Miami,
    'LA': count_value_LA,
    'Chicago': count_value_Chicago,
    'Chandler': count_value_Chandler}
df_counts = pd.DataFrame(counts, index=['Counts'])
df_counts
```

```python
#export cleaned data as csv
df.to_csv('Cleaned_Sold.csv', index=False)
```

```python
df_inflation = pd.read_csv('Cleaned_Inflation.csv')
df_inflation['StartDate'] = pd.to_datetime(df_inflation['StartDate'])
df_inflation['EndDate'] = pd.to_datetime(df_inflation['EndDate'])

# Convert date columns to YYYY-MM-DD format
df_inflation['StartDate'] = df_inflation['StartDate'].dt.strftime('%Y-%m-%d')
df_inflation['EndDate'] = df_inflation['EndDate'].dt.strftime('%Y-%m-%d')
df_inflation
```
